Convert a number base $\beta$ to a decimal number

$$
(a_na_{n-1}...a_1a_0.b_1b_2b_3...)_{\beta}
=
a_n \beta^n + a_{n-1} \beta^{n-1} + ... + a_1 \beta + a_0 + b_1 \beta^{-1} + b_2 \beta^{-2} + b_3 \beta^{-3} + ...
$$

Common Bases
* 02 Binary
* 08 Octal
* 10 Decimal
* 16 Hexadecimal
* 20 Vigesimal
* 60 Sexagesimal
